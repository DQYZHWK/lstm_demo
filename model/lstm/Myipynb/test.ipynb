```python
import csv
import jieba

# 打开 CSV 文件和输出文件
with open('./test.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    num = 0 
    for i,row in enumerate(reader):
        # 分词
        num +=1
        seg_list = jieba.cut(row[2])
        # 将分词结果转换为字符串，用空格隔开
        segmented_text = ' '.join(seg_list).rstrip('\n')

        # 写入到文本文件
        with open('lstm_test.txt', 'a', encoding='utf-8') as output_file:
            output_file.write(segmented_text + '\n')
    print(num)
```
